```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://github.com/Jeevita-R/ML_datasets/raw/refs/heads/main/House_price_prediction.csv')
df.head()
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, median_absolute_error
```

To load data from computer :-

df=pd.read_csv('House_price_prediction csv')

df.head()



```python
df.info()
```

```python
df = df.drop('Unnamed: 0',axis = 1) #axis = 0 -> row, axis = 1 -> col
df.head()
```

```python
df = pd.get_dummies(df, drop_first = True, dtype = int)
df.head()
```

```python
X = df.drop('Price', axis = 1)
y = df['Price']
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.3, random_state = 2)
```

```python
Lr_model = LinearRegression()
```

```python
Lr_model.fit(Xtrain, ytrain)
```

```python
y_pred_train = Lr_model.predict(Xtrain)
r2_score(ytrain, y_pred_train)
```

```python
y_pred_test = Lr_model.predict(Xtest)
r2_score(ytest, y_pred_test)
```

```python
Lr_model.score(Xtrain, ytrain)
```

```python
Lr_model.score(Xtest, ytest)
```

```python
R_model = Ridge(alpha=5)
R_model.fit(Xtrain, ytrain)
```

```python
print('Training efficiency:',R_model.score(Xtrain,ytrain))
print('Testing efficiency:',R_model.score(Xtest,ytest))
```

```python
L_model = Lasso(alpha = 1000)
L_model.fit(Xtrain, ytrain)
```

```python
print('Training efficiency:',L_model.score(Xtrain,ytrain))
print('Testing efficiency:',L_model.score(Xtest,ytest))
```
