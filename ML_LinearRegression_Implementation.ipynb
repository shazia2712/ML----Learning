```python
X = df.iloc[:,:-1]  # it will fetch all cols except last col
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, train_size=0.7, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
Lr = LinearRegression()
```

```python
Lr.fit(Xtrain, ytrain)
```

```python
Lr.coef_  # going to print the coefficient values (theta 1 value)
```

```python
Lr.intercept_   # this is intercept values (theta 0 value)
```

```python
pred = Lr.predict(Xtest)
pred
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
r2 = r2_score(ytest, pred)
r2
```

```python
s = X.shape
N = s[0]
P = s[1]
adgr2 = 1 - ((1-r2)*(N-1)/(N-P-1))
adgr2
```
